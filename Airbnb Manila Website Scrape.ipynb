```python
all_listings = []
for page_no in range(1, 16):
    if page_no == 1:
        url = "https://www.airbnb.com/s/Manila--Philippines/homes?place_id=ChIJi8MeVwPKlzMRH8FpEHXV0Wk&refinement_paths%5B%5D=%2Fhomes&tab_id=home_tab&flexible_trip_lengths%5B%5D=one_week&search_type=unknown&federated_search_session_id=166833c3-34ec-456f-af17-8d8a4fa1ef17&pagination_search=true"
    else:
        url = "https://www.airbnb.com/s/Manila--Philippines/homes?place_id=ChIJi8MeVwPKlzMRH8FpEHXV0Wk&refinement_paths%5B%5D=%2Fhomes&tab_id=home_tab&flexible_trip_lengths%5B%5D=one_week&search_type=unknown&federated_search_session_id=166833c3-34ec-456f-af17-8d8a4fa1ef17&pagination_search=true&items_offset=" + str(20 * (page_no - 1)) + "&section_offset=3"
    page = requests.get(url, headers={'User-agent': 'your bot 0.1'})
    soup = BeautifulSoup(page.content, "html.parser")
    listings = soup.find_all("div", class_="c4mnd7m dir dir-ltr")
    for i in range(len(listings)):
        data = listings[i].find("div", class_="g1tup9az cb4nyux dir dir-ltr")
        listing = []
        for child in data.children:
            listing.append(child.text)
        all_listings.append(listing)
    time.sleep(2)
```
